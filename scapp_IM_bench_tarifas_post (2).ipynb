```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

def clic(elem,d):
    d.execute_script("arguments[0].click();",elem)
    time.sleep(1)


edge_options = webdriver.EdgeOptions()
edge_options.add_argument('--headless')

driver= webdriver.Edge(options= edge_options)
driver.set_window_size(1920,1080)

url = 'https://www.movistar.com.pe/movil/postpago/planes-postpago'
driver.get(url)
time.sleep(3)

planes_lista =[]
planes = driver.find_elements(By.CSS_SELECTOR, 'div.swiper-wrapper')[1].find_elements(By.CSS_SELECTOR, 'div.swiper-slide')
print(len(planes))

for plan in planes:
    gigas = plan.find_element(By.CSS_SELECTOR, 'p.p-plan__slide__gigas').find_element(By.CSS_SELECTOR, 'span.p-plan__slide__cantidad')
    cantidad_gigas = driver.execute_script("return arguments[0].textContent;", gigas).replace(' ','')

    precio = plan.find_element(By.CSS_SELECTOR, 'p.p-plan__slide__precio').find_element(By.CSS_SELECTOR, 'span.p-plan__slide__cantidad')
    precio_local = driver.execute_script("return arguments[0].textContent;", precio).replace('S/ ','')

    precio_con_iva = round(float(precio_local),4)
    tasa_con_iva = 1.18
    precio_sin_iva = round(precio_con_iva/tasa_con_iva,4)
    tipo_de_cambio_respecto_al_dolar= 0.2682
    renta_en_dolares_sin_iva = round(precio_sin_iva*tipo_de_cambio_respecto_al_dolar,4)
    renta_en_dolares_con_iva = round(precio_con_iva*tipo_de_cambio_respecto_al_dolar,4)
    vigencia = '30 días'
    observaciones = " "
    plan_caract = ['Peru','Movistar','Postpago', cantidad_gigas, precio_sin_iva , tasa_con_iva, precio_con_iva, tipo_de_cambio_respecto_al_dolar, renta_en_dolares_sin_iva, renta_en_dolares_con_iva, vigencia, url, observaciones]
    planes_lista.append(plan_caract)

header_postpago = driver.find_element(By.CSS_SELECTOR, 'P.parr-planes-postpago__bajada')
driver.execute_script("arguments[0].scrollIntoView();", header_postpago)
time.sleep(1)
driver.save_screenshot('peru_movistar_postpago1.png')

siguiente = driver.find_element(By.XPATH, '/html/body/div[7]/div[2]/div[2]/div/div/div/section/div/div[2]/div/div/div/section[2]/div[2]/div[2]/div/div[3]')
siguiente.click()
time.sleep(1)
siguiente.click()
time.sleep(1)
driver.save_screenshot('peru_movistar_postpago2.png')

driver.quit()

columnas = ['País','Empresa','Válido para planes', 'Máximo de descarga', 'Precio o Renta Mensual En Moneda Local (Sin IVA)', '(1+IVA)', 'Precio o Renta Mensual En Moneda Local (Con IVA)', 'Tipo de Cambio Corriente (Respecto a US$)', 'Precio o Renta Mensual En Dólares, Sin IVA', 'Precio o Renta Mensual En Dólares Corrientes, Inc IVA', 'Vigencia','Link', 'Observaciones']
df = pd.DataFrame(planes_lista, columns= columnas)

df.to_excel('peru_movistar_tarifa_postpago.xlsx', index=False)
print('Proceso terminado con exito')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import os

ruta_carpeta= "D:\Postpago"

archivos_xlsx = [archivo for archivo in os.listdir(ruta_carpeta) if archivo.endswith('.xlsx')]

dataframes =[]

for archivo in archivos_xlsx:
    ruta_completa = os.path.join(ruta_carpeta,archivo)
    df = pd.read_excel(ruta_completa)
    dataframes.append(df)

df_final = pd.concat(dataframes, ignore_index=True)

print(df_final)

df_final.to_excel('tarifas_postpago.xlsx', index=False)
print('Proceso terminado con Exito!')
```
